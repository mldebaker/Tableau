```python
df2019_4thquarter = pd.concat([df201912, df201911, df201910, df201909, df201908, df201907,
df201906, df201905, df201904, df201903, df201902, df201901])
df2019_4thquarter
```

```python
df2019_4thquarter["Age"] = 2019 - df2019_4thquarter["birth year"]
df2019_4thquarter
```

```python
df2019_4thquarter = df2019_4thquarter.sort_index()
```

```python
df2019_4thquarter.to_csv('Data/4thquarter2019.csv', sep=',')
```

```python
df1 = df2019_4thquarter.loc[:, ['start station id', 
'start station name', 'start station latitude', 'start station longitude', 
'gender', 'Age', 'usertype', 'bikeid', 'tripduration', 'starttime']]
df1.head()
```

```python
df2 = df2019_4thquarter.loc[:, ['end station id', 
'end station name', 'end station latitude', 'end station longitude', 
'gender', 'Age',  'usertype', 'bikeid', 'tripduration', 'starttime']]
df2.head()
```

```python
df1.to_csv('Data/startlocations.csv', sep=',')
```

```python
df2.to_csv('Data/endlocatoins.csv', sep=',')
```

```python
df1.rename(columns
= {'start station id':'station id', 'start station name':'station name', 'start station latitude':'station latitude', 
'start station longitude':'station longitude'
}, inplace = True)
df1.head()
```

```python
df2.rename(columns
= {'end station id':'station id', 'end station name':'station name', 'end station latitude':'station latitude', 
'end station longitude':'station longitude'
}, inplace = True)
df2.head()
```

```python
location = pd.concat([df1, df2])
location.to_csv('Data/locations.csv', sep=',')
location.head()
```
